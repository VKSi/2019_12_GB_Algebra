(x - x1)/(x2 - x1) = (y - y1)/(y2 -y1) = (z - z1)/(z2 - z1) 
#### Как узнать, принадлежит прямая плоскости или нет?

### Ответ:

Так как через две различные точки в пространстве можно провести одну и только одну прямую, прямая будет принадлежать плоскости
в том и только в том случае, если две ее различные точки лежат на этой плоскости. То есть должны одновременно выполняться два условия:
##### A1*x1 + B1*y1 + C1*z1 + D1 = 0
##### A1*x2 + B1*y2 + C1*z2 + D1 = 0

### 5. Задание (в программе)

#### 1) Нарисуйте трехмерный график двух параллельных плоскостей.

### Ответ:

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
def par_surf(A=0, B=0, C=1, D=0, dx=0, dy=0, dz=1):
    fig = figure()
    ax = Axes3D(fig)
    X = np.arange(-5, 5, 0.5)
    Y = np.arange(-5, 5, 0.5)
    X, Y = np.meshgrid(X, Y)
    Z1 = -(A*X + B*Y + D)/C
    Z2 = -(A*(X - dx) + B*(Y - dy) + D)/C + dz
    ax.plot_wireframe(X, Y, Z1)
    ax.plot_wireframe(X, Y, Z2)
    show()
```

```python
par_surf(7, 3, 10, 2, 0, 0, 3)
```

#### 2) Нарисуйте трехмерный график двух любых поверхностей второго порядка.

### Ответ:

```python
def ellip_par(A=1, B=1):
    fig = figure()
    ax = Axes3D(fig)
    X = np.arange(-5, 5, 0.5)
    Y = np.arange(-5, 5, 0.5)
    X, Y = np.meshgrid(X, Y)
    Z = X**2/A**2 + Y**2/B**2
    ax.plot_surface(X, Y, Z)
    show()
```

```python
ellip_par(3,2)
```

```python
def hyp_par(A=1, B=1):
    fig = figure()
    ax = Axes3D(fig)
    X = np.arange(-5, 5, 0.5)
    Y = np.arange(-5, 5, 0.5)
    X, Y = np.meshgrid(X, Y)
    Z = X**2/A**2 - Y**2/B**2
    ax.plot_surface(X, Y, Z)
    show()
```

```python
hyp_par(5, 6)
```

#### 3. Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

### Ответ:

```python
def pol_segment(R1, ang1, R2, ang2):
    import numpy as np
    x = []
    y = []
    for i in range(101):
        x.append(R1 * np.cos(ang1) + (R2 * np.cos(ang2) - R1 * np.cos(ang1)) * i/100)
        y.append(R1 * np.sin(ang1) + (R2 * np.sin(ang2) - R1 * np.sin(ang1)) * i/100)
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axis('scaled')
```

### 4. Задание (в программе)

#### 1) Решите систему уравнений:
#### y = x ^ 2 – 1
#### exp(x) + x * (1 – y) = 1

### Ответ:

```python
x = np.linspace(-2, 3, 201)
plt.plot(x, x**2 -1)
plt.plot(x, 1 + (np.exp(x) - 1)/x)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return(y - x**2 + 1, np.exp(x) + x * (1 - y) - 1)
```

```python
x1, y1 = fsolve(equations, (-2, 1))
print(x1, y1)
```

```python
x1, y1 = fsolve(equations, (2, 1))
print(x1, y1)
```

#### 2) Решите систему уравнений и неравенств:

```python
def f1(x):
    return x**2 - 1

def f2(x, y):
    return np.exp(x) + x * (1 - y) - 1

x0 = fsolve(f1, -2)
if f2(x0, f1(x0)) > 0:
    print(x0)
    
x0 = fsolve(f1, 2)
if f2(x0, f1(x0)) > 0:
    print(x0)
```
